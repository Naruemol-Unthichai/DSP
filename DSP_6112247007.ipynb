```python
#ประกาศ
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display

```

```python
#Lab01
sin_sig = thinkdsp.SinSignal(freq=7007, amp=7, offset=0)
sin_sig.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#Lab02
#TriangleSignal(f)
#duration = signal.period*n
signal = thinkdsp.TriangleSignal(7)
duration = signal.period*70
segment = signal.make_wave(duration,framerate=10000)
segment.plot()

beamwave = signal.make_wave(duration, framerate=10000)
beamwave.apodize()
beamwave.make_audio()

```

```python
spectrum = beamwave.make_spectrum()
spectrum.plot()
```

```python
#ประกาศสี่เหลี่ยม
signal = thinkdsp.SquareSignal(700)
duration = signal.period*7
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)

wave = signal.make_wave(duration=0.5, framerate=10000)
wave.apodize()
wave.make_audio()
```

```python
spectrum = wave.make_spectrum()
spectrum.plot()
```
